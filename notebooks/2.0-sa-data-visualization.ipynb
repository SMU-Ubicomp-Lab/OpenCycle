```python
%pylab --no-import-all
%matplotlib inline
from os import path
import pandas as pd
import seaborn as sns
import plotly.offline as py
py.init_notebook_mode()
```

```python
file = path.join("..", "data", "interim", "df.csv")
df = pd.read_csv(file, index_col=0)
```

How many subjects?

```python
len(df.ID.drop_duplicates())
```

How many cycles per subject?

A subject is represented by an ID, and each row is a cycle, so we count the rows pertaining to each ID.

```python
sns.countplot(df.ID.value_counts());
```

How many segments per participant?
